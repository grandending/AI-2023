```python
import cv2
import os

def convert_to_binary_mask(image_path, save_path):
    # 이미지 불러오기 (흑백 이미지로)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # 픽셀 값이 0보다 큰 위치의 픽셀들을 하얀색으로 설정
    _, binary_mask = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY)

    # 결과 이미지 저장
    cv2.imwrite(save_path, binary_mask)

# 예시
image_dir = '/content/drive/Shareddrives/AI/validation_dataset/validation_mask'
output_dir = '/content/drive/Shareddrives/AI/validation_dataset/white_validation_mask'
os.makedirs(output_dir, exist_ok=True)

for file_name in os.listdir(image_dir):
    image_path = os.path.join(image_dir, file_name)
    save_path = os.path.join(output_dir, file_name)
    convert_to_binary_mask(image_path, save_path)
```

```python
# CSV 파일 불러오기
data = pd.read_csv('sorted_merged_label.csv')

# 데이터프레임 출력
data
```

```python
# 모든 파일 이름 가져오기
file_names = os.listdir(path+'training_image/')

# 정렬 함수 정의
def custom_sort(file_name):
    # 확장자 제외한 파일 이름만 추출
    base_name = os.path.splitext(file_name)[0]

    parts = base_name.split('_')
    # 파일 이름 형식이 맞지 않으면 정렬의 마지막에 위치시킵니다.
    if len(parts) != 4:
        return ('Z'*10, 999999)  # 임의의 큰 값

    sorting_key = parts[2]
    alpha_part = sorting_key[:1]
    numeric_part = int(sorting_key[1:])
    return (alpha_part, numeric_part)

# 정렬
sorted_file_names = sorted(file_names, key=custom_sort)

print(sorted_file_names)
print(len(sorted_file_names))
```

```python
import re
```

```python
# 이미지 폴더에서 파일 리스트 가져오기
file_names = sorted_file_names

# 정규 표현식 패턴 정의
pattern = r'^\d{2}_01_.*\.jpg$'

# 패턴에 맞는 파일만 선택
filtered_file_names = [f for f in file_names if re.match(pattern, f)]

sorted_file_names = filtered_file_names
```

```python
print(len(sorted_file_names))
```

```python
# 이미지 폴더에서 파일 리스트 가져오기
file_names = sorted_file_names

# 키와 몸무게를 저장할 리스트
heights = []
weights = []

# 파일명에서 모델 코드 추출 및 CSV에서 해당하는 행 찾기
for file_name in file_names:
    # 파일 확장자 확인
    if not file_name.endswith('.jpg'):
        continue

    # 파일명에서 모델 코드 추출 (예: "01_01_F051_04.jpg" -> "F051")
    model_code = file_name.split('_')[2]

    # CSV에서 모델 코드에 해당하는 행 찾기
    matching_rows = data[data.iloc[:, 2] == model_code]

    if not matching_rows.empty:
        # 키와 몸무게 가져오기
        height = matching_rows.iloc[0, 3]  # 3번째 열이 키
        weight = matching_rows.iloc[0, 35]  # 35번째 열이 몸무게

        # 키와 몸무게를 리스트에 저장
        heights.append(height)
        weights.append(weight)

        print(f"Model Code: {model_code}, Height: {height}, Weight: {weight}")
    else:
        print(f"No matching data found for Model Code: {model_code}")

# 확인용 출력
print("Heights:", heights)
print("Weights:", weights)
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 이미지 파일 경로를 포함한 전체 경로가 필요합니다.
# 예를 들어, 이미지 파일들이 'images' 폴더에 있다면, 아래와 같이 설정할 수 있습니다.
image_folder_path = path+"training_image/"

# 5x5 그리드로 이미지 표시
num_rows = 5
num_cols = 5

# 서브플롯을 위한 설정
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))

for i in range(num_rows):
    for j in range(num_cols):
        index = i * num_cols + j
        if index >= len(file_names):
            break

        # 이미지 로드
        img_path = os.path.join(image_folder_path, file_names[index])
        img = cv2.imread(img_path)

        if img is None:
            print(f"Unable to load image: {img_path}")
            continue

        # OpenCV는 이미지를 BGR로 로드하므로 RGB로 변환
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # 파일명에서 모델 코드 추출
        model_code = file_names[index].split('_')[2]

        # CSV에서 모델 코드에 해당하는 행 찾기
        matching_rows = data[data.iloc[:, 2] == model_code]

        if not matching_rows.empty:
            # 키와 몸무게 가져오기
            height = matching_rows.iloc[0, 3]  # 3번째 열이 키
            weight = matching_rows.iloc[0, 35]  # 35번째 열이 몸무게
            title = f"{file_names[index]}\nHeight: {height}, Weight: {weight}"
        else:
            title = f"{file_names[index]}\nNo matching data found"

        # 이미지와 키, 몸무게 정보 표시
        axes[i, j].imshow(img)
        axes[i, j].set_title(title)
        axes[i, j].axis('off')

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import models, transforms

import numpy as np
from fractions import Fraction
```

```python
img_paths = []
heights = []  # 키 값 리스트
weights = []  # 몸무게 값 리스트
processed_files = []  # 처리된 파일들을 추적하기 위한 리스트

for file_name in file_names:
    model_code = file_name.split('_')[2]
    matching_rows = data[data.iloc[:, 2] == model_code]

    if not matching_rows.empty:
        try:
            # 문자열이 분수 형태인지 확인하고, 분수 형태라면 Fraction을 사용하여 변환
            try:
                height = float(matching_rows.iloc[0, 3])
            except ValueError:
                height = float(Fraction(matching_rows.iloc[0, 3]))

            try:
                weight = float(matching_rows.iloc[0, 35])
            except ValueError:
                weight = float(Fraction(matching_rows.iloc[0, 35]))

            # 키와 몸무게 정보가 있을 경우만 리스트에 추가
            img_path = os.path.join(image_folder_path, file_name)
            img_paths.append(img_path)
            heights.append(height)
            weights.append(weight)
        except Exception as e:
            print(f"Error with file {file_name}: {e}")

    processed_files.append(file_name)

# img_paths에 포함되지 않은 파일을 찾기
missing_files = set(processed_files) - set([os.path.basename(path) for path in img_paths])

if missing_files:
    print("빠진 파일들:")
    for file in missing_files:
        print(file)
```

```python
def show_mask(mask, ax, random_color=False):
    if random_color:
        color = np.concatenate([np.random.random(3), np.array([0.6])], axis=0)
    else:
        color = np.array([30/255, 144/255, 255/255, 0.6])
    h, w = mask.shape[-2:]
    mask_image = mask.reshape(h, w, 1) * color.reshape(1, 1, -1)
    ax.imshow(mask_image)

def show_points(coords, labels, ax, marker_size=375):
    pos_points = coords[labels==1]
    neg_points = coords[labels==0]
    ax.scatter(pos_points[:, 0], pos_points[:, 1], color='green', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)
    ax.scatter(neg_points[:, 0], neg_points[:, 1], color='red', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)

def show_box(box, ax):
    x0, y0 = box[0], box[1]
    w, h = box[2] - box[0], box[3] - box[1]
    ax.add_patch(plt.Rectangle((x0, y0), w, h, edgecolor='green', facecolor=(0,0,0,0), lw=2))
```

```python
# 'training_mask' 폴더 생성 (이미 존재하지 않는 경우)
if not os.path.exists('back_training_mask'):
    os.mkdir('back_training_mask')

# img_paths에 있는 모든 이미지에 대하여 세그멘테이션 작업 진행
for img_path in img_paths:
    # 이미지 불러오기
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    predictor.set_image(image)

    input_box = np.array([700, 600, 1500, 2800])  # Box는 상황에 따라 조절이 필요할 수 있음
    masks, _, _ = predictor.predict(
        point_coords=None,
        point_labels=None,
        box=input_box[None, :],
        multimask_output=False,
    )

    # 예측된 세그멘테이션 마스크를 3차원으로 확장
    expanded_mask = np.repeat(masks[0][:, :, np.newaxis], 3, axis=2)

    # 원본 이미지에서 세그멘테이션 마스크에 해당하는 부분만 잘라냄
    segmented_region = np.where(expanded_mask, image, 0)

    # 잘라낸 영역을 'training_mask' 폴더에 저장
    # 파일명을 원본 파일명 기반으로 지정
    filename = os.path.basename(img_path).replace('.jpg', '_segmented.jpg')
    save_path = os.path.join('back_training_mask', filename)
    cv2.imwrite(save_path, segmented_region)
```
